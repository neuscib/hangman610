1. Description

**Hangman** is a version of the classic word-guessing game where the objective is to guess the name of a fruit before running out of attempts. The game randomly selects a fruit from a list of 5 options, and the player must guess the letters of the fruit within a limited number of tries. If you guess the letter correctly, it appears in the right place; if not, you lose an attempt.


#### What I learned
Throughout this project, I improved my skills in:
- Writing functions to modularize code
- Using `input()` to interact with the user
- Using conditionals and string methods (`isalpha()`, `lower()`)
- Validating user input
- Randomly selecting items from a list

3. Usage instructions

- Run the game: Once you have cloned the repository, you can run the hangman.py file, which contains the main game code. Use the following command in your terminal: 

python hangman.py

- Play the game: The game will prompt you to enter a letter. If the letter is correct, it will be shown in the correct place. If the letter is incorrect, you will lose an attempt. The goal is to guess the fruit before running out of attempts.

- Exit the game: If you wish to exit the game at any point, simply close the program with Ctrl+C or terminate the process in your terminal.
